```python
#importing dependencies
import pandas as pd
import numpy as np
import datetime
```

```python
# File to Load 
listings_resources = "listings.csv"
```

```python
# Read the csv and store into Pandas DataFrames
listings_resources_2 = pd.read_csv(listings_resources)
listings_resources_2
```

```python
#Filtering the data based on latitudes and longitudes
listings_filtered = listings_resources_2.loc[(listings_resources_2['latitude']>= -32.5468) & (listings_resources_2['latitude'] <= -31.7725) & (listings_resources_2['longitude'] <= 116.0593)]
listings_filtered
```

```python
#Filtering the data based on number of nights (Number of nights <= 15)
listings_filtered_nights = listings_filtered.loc[(listings_filtered['minimum_nights'] <= 15)]
listings_filtered_nights
```

```python
#Removing unnecessary columns
listings_filtered_nights_columns = listings_filtered_nights.drop(['neighbourhood_group','license'], axis=1)
listings_filtered_nights_columns
```

```python
#Dropping the rows with no values in the last review
listings_filtered_nights_columns_droppedna = listings_filtered_nights_columns[listings_filtered_nights_columns['last_review'].notna()]
listings_filtered_nights_columns_droppedna
```

```python
#File to load
reviews = "reviews.csv"
# Read the csv and store into Pandas DataFrames
reviews_2 = pd.read_csv(reviews, low_memory=False)
reviews_2
```

```python
#merging the dataset
#review_data_merged = pd.merge(reviews_2, listings_resources_2, how='left', on= ['listing_id','id'])
review_data_merged = pd.merge(left=reviews_2, right=listings_resources_2, left_on='listing_id', right_on='id', how='left')
review_data_merged.to_csv('review_data_merged.csv', encoding='utf-8')
review_data_merged
```

```python
#Filtering the data based on latitudes and longitudes
review_data_merged_filtered = review_data_merged.loc[(review_data_merged['latitude']>= -32.5468) & (review_data_merged['latitude'] <= -31.7725) & (review_data_merged['longitude'] <= 116.0593)]
review_data_merged_filtered
```

```python
#Filtering the data based on number of nights (Number of nights <= 15)
review_data_merged_filtered_nights = review_data_merged_filtered.loc[(review_data_merged_filtered['minimum_nights'] <= 15)]
review_data_merged_filtered_nights
```

```python
#Removing unnecessary columns
review_data_merged_filtered_nights_unncols = review_data_merged_filtered_nights.drop(['neighbourhood_group','license'], axis=1)
review_data_merged_filtered_nights_unncols
```

```python
#Adding column to dataframe
review_data_merged_filtered_nights_unncols['Count before and after Covid'] = 'None'
review_data_merged_filtered_nights_unncols
```

```python
#Changing the date format
review_data_merged_filtered_nights_unncols['date'] = pd.to_datetime(review_data_merged_filtered_nights_unncols.date, infer_datetime_format=True)
review_data_merged_filtered_nights_unncols
```

```python
#Assigning value to the column 'Before Covid'
review_data_merged_filtered_nights_unncols.loc[(review_data_merged_filtered_nights_unncols['date'] > '2016-01-01') & (review_data_merged_filtered_nights_unncols['date'] < '2022-03-11'), 'Count before and after Covid'] = 0 
review_data_merged_filtered_nights_unncols.loc[(review_data_merged_filtered_nights_unncols['date'] > '2020-03-11') & (review_data_merged_filtered_nights_unncols['date'] < '2022-03-03'), 'Count before and after Covid'] = 2
review_data_merged_filtered_nights_unncols.loc[review_data_merged_filtered_nights_unncols['date'] >= '2022-03-03', 'Count before and after Covid'] = 4 

review_data_merged_filtered_nights_unncols
#reviews_2.loc[reviews_2['date'] <= '2020-03-11', 'Count before and after Covid'] = 0 
#reviews_2.loc[(reviews_2['date'] > '2020-03-11') & (reviews_2['date'] < '2022-03-03'), 'Count before and after Covid'] = 2
#reviews_2.loc[reviews_2['date'] >= '2022-03-03', 'Count before and after Covid'] = 4 

#reviews_2
```

```python
#Segregating before and after covid
bins = [0, 1, 3,4]
group_names = ["Before Covid","Covid","After Covid"]
review_data_merged_filtered_nights_unncols["Count before and after Covid"] = pd.cut(review_data_merged_filtered_nights_unncols["Count before and after Covid"], bins, labels=group_names, include_lowest=True)
review_data_merged_filtered_nights_unncols
```

```python
review_data_merged_filtered_nights_unncols = review_data_merged_filtered_nights_unncols.groupby(['listing_id', 'Count before and after Covid'])
review_data_merged_filtered_nights_unncols
```
